```python
from transformers import pipeline
import pandas as pd
```

```python
tqa = pipeline(task="table-question-answering",model="google/tapas-base-finetuned-wtq")
```

```python
table=pd.read_csv("/content/jira final data.csv")
```

```python
table = table.astype(str)
```

```python
table
```

```python
query = "Give me any 3 Issue key for issues that has status: Closed"
print(tqa(table=table,query=query)["answer"])
```

```python
query = "issues that has status: Closed"
print(tqa(table=table,query=query)["answer"])
```
